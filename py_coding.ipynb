```python
# Find all pairs that sum to a target value
nums = [1,2,3,4,5,6]
target = 7
pairs = []

for i in range(len(nums)):
    for j in range(i+1, len(nums)):
        if nums[i] + nums[j] == target:
            pairs.append((nums[i], nums[j]))

print(pairs)
```

```python
# 1. Keep only events from user_id in {2,3}
# 2. Filter to timestamps between 2024-01-01 02:00 and 2024-01-01 08:00

import pandas as pd

df = pd.DataFrame({
    "timestamp": pd.date_range("2024-01-01", periods=10, freq="h"),
    "user_id":   [1,1,2,2,2,3,3,3,3,3],
    "value":     [5,7,2,3,9,1,4,5,8,6]
})
df
```

```python
df1 = df[(df['user_id']==2) | (df['user_id']==3)] # bitwise
df1
```

```python
df2 = df[(df['timestamp'] > "2024-01-01 02:00") &
    (df['timestamp'] < "2024-01-01 08:00")] # bitwise
df2
```

```python
# 3. Resample the data in 1-hour buckets and compute the mean value per hour for each user.
df_indexed = df.set_index("timestamp")

step1 = (
    df_indexed.groupby("user_id")["value"].resample("1H").mean()
)
step1
```

```python
step1_df = step1.reset_index().sort_values(["user_id", "timestamp"])
step1_df
```

```python
# After resampling, calculate a 3-hour rolling mean for each user.

step2 = (
    step1_df.groupby("user_id").rolling("3H",on="timestamp")["value"].mean()
    .rename("rolling_3h_mean")
    .reset_index()
)
step2
```
